### to check missing values

```python
missing  = df.isnull().sum()
missing = missing[missing > 0].sort_values(ascending=False)
print(missing)
```

### to check target distribution- is it skewed?

```python
print(df['SalePrice'].describe())

plt.figure(figsize=(8,4))
plt.hist(df['SalePrice'], bins=50, color='steelblue', edgecolor='white')
plt.xlabel('Sale Price ($)')
plt.ylabel('Count')
plt.title('Distribution of House Sale Prices')
plt.show()
```

### from above operations we identified that Pool QC, Misc Feature and Alley data is mostly missing so we remove it

```python
# Order and PID are just row IDs, not features
# Pool QC, Misc Feature, Alley are mostly empty (90%+ missing)
cols_to_drop = ['Order', 'PID', 'Pool QC', 'Misc Feature', 'Alley']

df = df.drop(columns=cols_to_drop)
print("Remaining columns:", df.shape[1])
```
