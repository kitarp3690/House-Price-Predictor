### from above operations we identified that Pool QC, Misc Feature and Alley data is mostly missing so we remove it

```python
# Order and PID are just row IDs, not features
# Pool QC, Misc Feature, Alley are mostly empty (90%+ missing)
cols_to_drop = ['Order', 'PID', 'Pool QC', 'Misc Feature', 'Alley']

df = df.drop(columns=cols_to_drop)
print("Remaining columns:", df.shape[1])
```

### filling categorical missing values with 'None'

```python
# For these columns, missing simply means the house doesn't have that feature
# e.g. no fireplace = Fireplace Qu is Nan, we fill it with None
fill_none_cols = [
    'Fence', 'Fireplace Qu', 'Mas Vnr Type', 'Garage Type', 'Garage Qual', 'Garage Cond', 
    'Bsmt Qual', 'Bsmt Cond', 'Bsmt Exposure',
    'BsmtFin Type 1', 'BsmtFin Type 2'
]

for col in fill_none_cols:
    df[col] = df[col].fillna('None')
print('Done filling categorical nulls')
```

### filling numeric missing values

```python
# Lot Frontage: fill with median (most neutral choice)
df['Lot Frontage'] = df['Lot Frontage'].fillna(df['Lot Frontage'].median())

# Mas Vnr Area: no masonry = 0 sqrft
df['Mas Vnr Area'] = df['Mas Vnr Area'].fillna(0)

# Garage year built: no garage = 0
df['Garage Yr Blt'] = df['Garage Yr Blt'].fillna(0)

# Basement numeric cols: no basement = 0 sqft
bsmt_num_cols = ['BsmtFin SF 1', 'BsmtFin SF 2', 'Bsmt Unf SF',
                 'Total Bsmt SF', 'Bsmt Full Bath', 'Bsmt Half Bath']
for col in bsmt_num_cols:
    df[col] = df[col].fillna(0)

# Garage numeric cols: no garage = 0
df['Garage Cars'] = df['Garage Cars'].fillna(0)
df['Garage Area'] = df['Garage Area'].fillna(0)

print("Done filling numeric nulls")
```

### filling remaining 1-row missing values

```python
# Electrical: just 1 missing - fill with most common value
df['Electrical'] = df['Electrical'].fillna(df['Electrical'].mode()[0])
print('Done')
```

### verify no missing values remain

```python
remaining = df.isnull().sum()
remaining = remaining[remaining > 0]

if len(remaining) == 0:
    print('No misssing values remaining!')
else:
    print('Still missing:')
    print(remaining)
```

### fixing the Skewed ScalePrice with log transform

```python
import numpy as np

# Log transform makes the distribution normal — better for linear regression
df['SalePrice'] = np.log(df['SalePrice'])

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.hist(df['SalePrice'], bins=50, color='steelblue', edgecolor='white')
plt.title('SalePrice after log transform')
plt.xlabel('log(SalePrice)')

plt.tight_layout()
plt.show()

print("SalePrice stats after log transform:")
print(df['SalePrice'].describe())
```

```python
# Age of house when sold
df['House Age'] = df['Yr Sold'] - df['Year Built']

# Years since last remodel
df['Remod Age'] = df['Yr Sold'] - df['Year Remod/Add']

# Total square footage (basement + above ground)
df['Total SF'] = df['Total Bsmt SF'] + df['Gr Liv Area']

# Total bathrooms (full bath counts as 1, half bath as 0.5)
df['Total Baths'] = (df['Full Bath'] + df['Bsmt Full Bath'] +
                     0.5 * (df['Half Bath'] + df['Bsmt Half Bath']))

# Does the house have a garage? (1 = yes, 0 = no)
df['Has Garage'] = (df['Garage Area'] > 0).astype(int)

# Does the house have a fireplace?
df['Has Fireplace'] = (df['Fireplaces'] > 0).astype(int)

# Does the house have a basement?
df['Has Basement'] = (df['Total Bsmt SF'] > 0).astype(int)

print("New features created:")
print(df[['House Age', 'Remod Age', 'Total SF',
          'Total Baths', 'Has Garage',
          'Has Fireplace', 'Has Basement']].head())
```

```python
# We pick the most impactful numeric + engineered features
# Dropping raw columns we've already captured in engineered ones

features = [
    # Size features
    'Total SF', 'Gr Liv Area', '1st Flr SF', '2nd Flr SF',
    'Total Bsmt SF', 'Garage Area', 'Lot Area',

    # Quality features
    'Overall Qual', 'Overall Cond',

    # Age features
    'House Age', 'Remod Age',

    # Room features
    'Total Baths', 'Bedroom AbvGr', 'TotRms AbvGrd', 'Fireplaces',

    # Binary features
    'Has Garage', 'Has Fireplace', 'Has Basement',
]

X = df[features]
y = df['SalePrice']

print("X shape:", X.shape)
print("y shape:", y.shape)
print("\nAny missing in X?", X.isnull().sum().sum())
```

```python
# See which features are most correlated with price
correlations = X.join(y).corr()['SalePrice'].drop('SalePrice')
correlations = correlations.sort_values(ascending=False)

print(correlations)

# Plot it
plt.figure(figsize=(8, 6))
correlations.sort_values().plot(kind='barh', color='steelblue')
plt.title('Feature correlation with SalePrice')
plt.xlabel('Correlation')
plt.tight_layout()
plt.show()
```
